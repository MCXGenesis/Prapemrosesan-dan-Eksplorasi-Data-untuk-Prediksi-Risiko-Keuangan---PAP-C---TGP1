HASIL PERBANDINGAN VISUALISASI APP_TRAIN DENGAN APP_TEST

```python
plt.figure(figsize=(12, 5))

# Histogram DAYS_EMPLOYED untuk app_train
plt.subplot(1, 2, 1)
sns.histplot(app_train["DAYS_EMPLOYED"], bins=30, kde=True, color="blue")
plt.title("Distribusi DAYS_EMPLOYED - Train")
plt.xlabel("DAYS_EMPLOYED")
plt.ylabel("Frekuensi")

# Histogram DAYS_EMPLOYED untuk app_test
plt.subplot(1, 2, 2)
sns.histplot(app_test["DAYS_EMPLOYED"], bins=30, kde=True, color="red")
plt.title("Distribusi DAYS_EMPLOYED - Test")
plt.xlabel("DAYS_EMPLOYED")
plt.ylabel("Frekuensi")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

# Distribusi OCCUPATION_TYPE untuk Train
plt.subplot(1, 2, 1)
sns.countplot(y="OCCUPATION_TYPE", data=app_train, order=app_train["OCCUPATION_TYPE"].value_counts().index, palette="coolwarm")
plt.title("Distribusi OCCUPATION_TYPE - Train")
plt.xlabel("Frekuensi")
plt.ylabel("OCCUPATION_TYPE")

# Distribusi OCCUPATION_TYPE untuk Test
plt.subplot(1, 2, 2)
sns.countplot(y="OCCUPATION_TYPE", data=app_test, order=app_train["OCCUPATION_TYPE"].value_counts().index, palette="coolwarm")
plt.title("Distribusi OCCUPATION_TYPE - Test")
plt.xlabel("Frekuensi")
plt.ylabel("")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

# Distribusi NAME_INCOME_TYPE untuk Train
plt.subplot(1, 2, 1)
sns.countplot(y="NAME_INCOME_TYPE", data=app_train, order=app_train["NAME_INCOME_TYPE"].value_counts().index, palette="coolwarm")
plt.title("Distribusi NAME_INCOME_TYPE - Train")
plt.xlabel("Frekuensi")
plt.ylabel("NAME_INCOME_TYPE")

# Distribusi NAME_INCOME_TYPE untuk Test
plt.subplot(1, 2, 2)
sns.countplot(y="NAME_INCOME_TYPE", data=app_test, order=app_train["NAME_INCOME_TYPE"].value_counts().index, palette="coolwarm")
plt.title("Distribusi NAME_INCOME_TYPE - Test")
plt.xlabel("Frekuensi")
plt.ylabel("")

plt.tight_layout()
plt.show()
```
