VISUALISASI VARIABEL ACCEPTED

```python
# Visualisasi distribusi variabel target (loan_status)
plt.figure(figsize=(10,4))
sns.countplot(y="loan_status", data=accepted, palette='coolwarm',
              order=accepted['loan_status'].value_counts().index)
plt.title('Distribusi Variabel Loan Status')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(accepted["dti"], bins=30, color="crimson")
plt.title("Distribusi Variabel DTI")
plt.xlabel("DTI")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# Visualisasi distribusi variabel int_rate
plt.figure(figsize=(10, 5))
sns.histplot(data=accepted, x="int_rate", bins=50, color="green")
plt.title("Distribusi Suku Bunga (Interest Rate)")
plt.xlabel("Interest Rate (%)")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# Visualisasi distribusi variabel loan_amnt
plt.figure(figsize=(10, 5))
sns.histplot(data=accepted, x="loan_amnt", bins=50,  color="blue")
plt.title("Distribusi Loan Amount")
plt.xlabel("Loan Amount")
plt.ylabel("Frekuensi")
plt.show()
```

VISUALISASI VARIABEL REJECTED

```python
#Bar Chart Employment Length
plt.figure(figsize=(10, 5))
sns.countplot(data=rejected, x="Employment Length", palette="coolwarm")
plt.title("Distribusi Employment Length pada Pemohon Ditolak")
plt.xlabel("Employment Length")
plt.ylabel("Jumlah")
plt.xticks(rotation=45)
plt.show()
```

```python
#Histogram Jumlah Pinjaman yang Diminta
plt.figure(figsize=(10, 5))
sns.histplot(data=rejected, x="Amount Requested", bins=30, color="blue")
plt.title("Distribusi Amount Requested")
plt.xlabel("Amount Requested")
plt.ylabel("Frekuensi")
plt.show()
```

```python
#Bar Chart Distribusi Berdasarkan State
plt.figure(figsize=(12, 6))
sns.countplot(data=rejected, x="State", order=rejected["State"].value_counts().index, palette="magma")
plt.title("Distribusi Penolakan Berdasarkan State")
plt.xlabel("State")
plt.ylabel("Jumlah")
plt.xticks(rotation=90)
plt.show()
```

LIHAT PERBANDINGAN DATASET ACCEPTED DAN REJECTED

```python
#Distribusi Jumlah Pinjaman yang Diminta (Loan Amount vs Amount Requested)
plt.figure(figsize=(12, 5))

# Visualisasi distribusi variabel loan_amnt
plt.subplot(1, 2, 1)
sns.histplot(data=accepted, x="loan_amnt", bins=50,  color="blue")
plt.title("Distribusi Loan Amount")
plt.xlabel("Loan Amount")
plt.ylabel("Frekuensi")

#Histogram Jumlah Pinjaman yang Diminta
plt.subplot(1, 2, 2)
sns.histplot(data=rejected, x="Amount Requested", bins=30, color="blue")
plt.title("Distribusi Amount Requested")
plt.xlabel("Amount Requested")
plt.ylabel("Frekuensi")

plt.tight_layout()
plt.show()
```
